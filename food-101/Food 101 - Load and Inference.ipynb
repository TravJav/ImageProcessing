```python
from fastai import *
from fastai.vision import *
```

```python
pred_model = load_learner(path='./', file='food-101.pkl') # Load in model
```

```python
# Load in sample image
path = Path('./food-101/images/apple_pie/1005649.jpg')
img = open_image(path)
```

```python
img.show(figsize=(12,12))
```

```python
label, _, _ = pred_model.predict(img)
```

```python
label = str(label) # Convert to string so that it's serializable in case you want to send a JSON response
print(label)
```
